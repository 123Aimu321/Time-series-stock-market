```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('nifty50/NIFTY50_all.csv')

# Preview the data
print("Initial shape:", df.shape)
print(df.head())
```

```python
# Step 1: Parse Date column
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
# Step 2: Drop rows with invalid dates
df = df.dropna(subset=['Date'])
```

```python
# Step 3: Sort by Date
df = df.sort_values('Date').reset_index(drop=True)
```

```python
# Step 4: Handle missing values
# Option 1: Forward fill
df.fillna(method='ffill', inplace=True)

# Option 2: Drop remaining nulls (if any)
df.dropna(inplace=True)
```

```python
# Step 5: Set Date as index (optional but useful for time series)
df.set_index('Date', inplace=True)
```

```python
# Step 6: Filter columns (if needed)
# Keep only relevant columns for forecasting
columns_to_keep = ['Open', 'High', 'Low', 'Close', 'Volume']
df = df[columns_to_keep]
```

```python
# Step 7: Feature Engineering (optional)
# Add daily returns
df['Daily_Return'] = df['Close'].pct_change()

# Add rolling averages
df['MA_7'] = df['Close'].rolling(window=7).mean()
df['MA_30'] = df['Close'].rolling(window=30).mean()
```

```python
# Final check
print("Processed shape:", df.shape)
print(df.tail())
```

```python
# Save cleaned data
df.to_csv('nifty50_cleaned.csv')
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
# Scale data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df[['Close']])
```

```python
# Create sequences
X, y = [], []
for i in range(60, len(scaled_data)):
    X.append(scaled_data[i-60:i])
    y.append(scaled_data[i])
X, y = np.array(X), np.array(y)
```

```python
# Reshape
X = X.reshape((X.shape[0], X.shape[1], 1))
```

```python
# Build model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X.shape[1], 1)))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.fit(X, y, epochs=20, batch_size=32)

```

```python
# Forecast next 30 days
last_60 = scaled_data[-60:]
input_seq = last_60.reshape((1, 60, 1))
predictions = []
for _ in range(30):
    pred = model.predict(input_seq)[0][0]
    predictions.append(pred)
    input_seq = np.append(input_seq[:,1:,:], [[[pred]]], axis=1)
```

```python
# Inverse scale
forecast = scaler.inverse_transform(np.array(predictions).reshape(-1,1))
```

```python
# Plot
plt.plot(df['Close'], label='Actual')
plt.plot(pd.date_range(df.index[-1], periods=30, freq='B'), forecast, label='LSTM Forecast')
plt.legend()
plt.show()
```

```python
#📏 Model Evaluation Techniques for Time Series Forecasting
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def evaluate_model(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    r2 = r2_score(y_true, y_pred)

    print(f"MAE: {mae:.2f}")
    print(f"MSE: {mse:.2f}")
    print(f"RMSE: {rmse:.2f}")
    print(f"MAPE: {mape:.2f}%")
    print(f"R² Score: {r2:.2f}")
```

```python
# Get actual values for comparison
y_true = df['Close'][-30:]  # last 30 actual values
y_pred = model_fit.predict(start=len(df)-30, end=len(df)-1)

evaluate_model(y_true, y_pred)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Fit SARIMA model
sarima_model = SARIMAX(df['Close'], order=(1,1,1), seasonal_order=(1,1,1,12))
sarima_fit = sarima_model.fit()

# Forecast last 30 steps for evaluation
y_pred_sarima = sarima_fit.predict(start=len(df)-30, end=len(df)-1)
y_true = df['Close'][-30:]

evaluate_model(y_true, y_pred_sarima)
```

```python
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense

# Scale data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df[['Close']])

# Create sequences
X, y = [], []
for i in range(60, len(scaled_data)):
    X.append(scaled_data[i-60:i])
    y.append(scaled_data[i])
X, y = np.array(X), np.array(y)

# Train-test split
X_train, X_test = X[:-30], X[-30:]
y_train, y_test = y[:-30], y[-30:]

# Build model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X.shape[1], 1)))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.fit(X_train, y_train, epochs=20, batch_size=32)

# Predict
y_pred_lstm = model.predict(X_test)
y_pred_lstm = scaler.inverse_transform(y_pred_lstm)
y_true_lstm = scaler.inverse_transform(y_test.reshape(-1,1))

evaluate_model(y_true_lstm, y_pred_lstm)
```
