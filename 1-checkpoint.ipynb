```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

```python
%matplotlib inline
sns.set_style("whitegrid")
```

```python
DATA_PATH = "nifty50/NIFTY50_all.csv"

df = pd.read_csv(
    DATA_PATH,
    parse_dates=["Date"],
    index_col="Date"
)

print(f"Rows Ã— Columns: {df.shape}")
df.head()
```

```python
# 3.1 Columns and dtypes
print(df.dtypes)
```

```python
# 3.2 Unique tickers
tickers = df["Symbol"].unique()
print(f"Loaded {len(tickers)} tickers:" )
```

```python
# 3.3 Missing values overall
total_missing = df.isna().sum().sum()
print(f"Total missing cells: {total_missing}")
```

```python
# 3.4 Missing values by column
print(df.isna().sum().sort_values(ascending=False))
```

```python
# 3.5 Visualize missingness (per column)
msno.matrix(df, figsize=(12, 4))
plt.title("Missing Data Matrix")
plt.show()
```

```python
# Pivot to have tickers as columns
close_df = df.pivot_table(values="Close", index=df.index, columns="Symbol")
```

```python
# Compute equal-weight index
ew_index = close_df.mean(axis=1)
ew_index.name = "EW_Index"

```

```python
# Quick plot
plt.figure(figsize=(14, 5))
ew_index.plot(color="black")
plt.title("Equal-Weight Proxy for NIFTY-50 (Average Close Price)")
plt.ylabel("Price (INR)")
plt.xlabel("Date")
plt.show()
```

```python
selected = ["RELIANCE", "TCS", "HDFCBANK"]

plt.figure(figsize=(14, 6))
for sym in selected:
    series = close_df[sym]
    plt.plot(series, label=sym)

plt.title("Closing Prices: Selected Tickers")
plt.xlabel("Date")
plt.ylabel("Price (INR)")
plt.legend()
plt.show()
```

```python
# 1. Create an indexed DataFrame (base 100 at first observation)
indexed = close_df[selected] / close_df[selected].iloc[0] * 100

# 2. Plot the normalized series
plt.figure(figsize=(14, 6))
for sym in selected:
    plt.plot(indexed[sym], label=sym)
plt.title("Relative Performance (Indexed to 100 at Start)")
plt.xlabel("Date")
plt.ylabel("Indexed Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

def compute_stats(price):
    returns = price.pct_change().dropna()
    total_ret = (price.iloc[-1] / price.iloc[0] - 1) * 100
    days = (price.index[-1] - price.index[0]).days
    cagr = (price.iloc[-1] / price.iloc[0]) ** (365/days) - 1
    max_dd = (price / price.cummax() - 1).min() * 100
    ann_vol = returns.std() * np.sqrt(252) * 100
    sharpe = (returns.mean() * 252) / (returns.std() * np.sqrt(252))
    return total_ret, cagr*100, max_dd, ann_vol, sharpe

stats = {sym: compute_stats(close_df[sym]) for sym in selected}
stats_df = pd.DataFrame(stats, index=[
    "Total Return (%)", "CAGR (%)", "Max Drawdown (%)",
    "Annualized Vol (%)", "Sharpe Ratio"
]).T
close_df[selected].isna().sum()
clean_df = close_df[selected].ffill().bfill()
stats = {sym: compute_stats(clean_df[sym]) for sym in selected}
stats_df = pd.DataFrame(stats, index=[
    "Total Return (%)", "CAGR (%)", "Max Drawdown (%)",
    "Annualized Vol (%)", "Sharpe Ratio"
]).T.round(2)
print(stats_df)
```

```python
returns = clean_df.pct_change()

rolling_mean = returns.rolling(30).mean() * 252
rolling_vol  = returns.rolling(30).std() * (252**0.5)

fig, axes = plt.subplots(2, 1, figsize=(14, 8), sharex=True)

rolling_mean.plot(ax=axes[0], title="30-Day Rolling Mean Returns (Annualized)")
axes[0].grid(True)

rolling_vol.plot(ax=axes[1], title="30-Day Rolling Volatility (Annualized)")
axes[1].grid(True)

plt.xlabel("Date")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

log_rets = np.log(clean_df / clean_df.shift(1)).dropna()
corr = log_rets.corr()

plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix (Log Returns)")
plt.show()
```
